```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python
data = pd.read_csv('franken_v2.99_10yrs__vatiras_granvik_10k_obs_streaked.txt', sep=' ')
```

```python
data['night'].values[ack]
```

```python
ids = data['obj_id'].values
night = data['night'].values
```

```python
n_n = []
for nid in np.unique(ids):
    good = np.where(ids == nid)[0]
    n_n.append(np.size(np.unique(night[good])))
```

```python
_temp = plt.hist(n_n, bins=np.arange(np.max(n_n)+2)-0.5)
plt.xlabel('Number of unique nights')
plt.ylabel('N unique obj_id')
plt.title('Vatiras after streak removal')
```
